## Popularity based recommendation system

```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name.head()
```

```python
num_of_ratings = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index().sort_values(by='Book-Rating' ,ascending=False)
num_of_ratings.rename(columns={'Book-Rating':'no_ratings' }, inplace = True)

avg_ratings = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

popular_books = num_of_ratings.merge(avg_ratings, on = 'Book-Title')

popular_books
```

```python
popular_books = popular_books[popular_books['no_ratings'] > 250]
top_50 = popular_books.sort_values(by='Book-Rating', ascending=False).head(50).reset_index(drop=True )
top_50.head()
```

```python
top_50 = top_50.merge(books, on='Book-Title', how = 'left').drop_duplicates('Book-Title')
```

```python
top_50.head()
```

```python
top_50 = top_50[['Book-Title','Book-Author','Book-Rating','Image-URL-M','no_ratings']]
top_50.shape
```

## Collaborative filtering based system

```python
ratings_with_name.head()
```

```python
x = ratings_with_name.groupby('User-ID').agg({'Book-Rating':'count'}).sort_values(ascending=False, by = 'Book-Rating')
valid_users = x[x['Book-Rating'] > 200].index
valid_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(valid_users)]
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').agg({'Book-Rating':'count'})
y = y[y['Book-Rating'] >= 50].index
filtered_rating =  filtered_rating[filtered_rating['Book-Title'].isin(y)]
filtered_rating
```

```python
pt = filtered_rating.pivot_table(index='Book-Title',columns='User-ID', values='Book-Rating')
```

```python
pt.head()
```

```python
pt.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
